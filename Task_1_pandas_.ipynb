# Name the students with passmark

```python
students_pass_mark = df_stumar[df_stumar['Marks'] < 25].merge(df_stu, left_on='Student ID', right_on='Student ID')['Name'].unique()
students_pass_mark
```

```python
def students_pass_mark_subject(subject):
    pass_students = df_stumar[(df_stumar['Marks'] = 25) & (df_stumar['Subject'] == subject)]
    pass_students_info = pass_students.merge(df_stu, left_on='Student ID', right_on='Student ID')
    pass_students_names = pass_students_info['Name'].unique()
    return pass_students_names

# Example usage:
pass_students_math = students_pass_mark_subject('Mathematics')
pass_students_science = students_pass_mark_subject('Science')
pass_students_comp=students_pass_mark_subject('Computer science')
pass_students_soci=students_pass_mark_subject('Sociology')
pass_students_acc=students_pass_mark_subject('Accountancy')
pass_students_eco=students_pass_mark_subject('Economics')
print(pass_students_math)
print(pass_students_science)
print(pass_students_comp)
print(pass_students_soci)
print(pass_students_acc)
print(pass_students_eco)
```
